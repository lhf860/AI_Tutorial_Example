```python


imagetype2columns = {'住院发票': ['性别', '入院日期', '出院日期', '住院天数', '票据代码', '票据号码', '校验码', '开票日期', '收款单位', '收款人', '复核人', '医疗机构类型', '医保类型', '合计金额（大写）', '（小写）', '诊查费', '检查费', '化验费', '治疗费', '手术费', '卫生材料费', '西药费', '中药饮片', '中成药费', '一般诊疗费', '床位费', '护理费', '挂号费', '其他收费项目', '医保统筹基金支付', '统筹支付', '个人现金支付', '个人账户支付', '个人自付', '其他支付', '自付一', '自付二'],
                     '购药发票': ['票据代码', '票据号码', '校验码', '开票日期', '收款人', '复核人', '价税合计（大写）', '（小写）'], 
                     '门诊发票': ['性别', '票据代码', '票据号码', '校验码', '开票日期', '收款单位', '收款人', '复核人', '医疗机构类型', '医保类型', '合计金额（大写）', '（小写）', '诊查费', '检查费', '化验费', '治疗费', '手术费', '卫生材料费', '西药费', '中药饮片', '中成药费', '一般诊疗费', '床位费', '护理费', '挂号费', '其他收费项目', '医保统筹基金支付', '统筹支付', '个人现金支付', '个人账户支付', '个人自付', '其他支付', '自付一', '自付二'], 
                     '出院小结': ['性别', '年龄', '医院名称', '组织机构代码', '医疗机构类型', '入院日期', '出院日期', '住院天数']}

image_column_describle = {"住院发票": ""}


"""

生成用于微调的prompt
"""


prompt_template = """这是一张#住院发票#的照片, 请返回下面字段的值,注意：如果图片中未出现该字段，返回‘无’；若图片中出现该字段但没有对应值，返回‘/’：
性别：候选值为男、女、无
入院日期：该字段与住院日期同义，如果不存在这两个字段，则不取住院天数：gaizi'duanyuzhu'yuanri'shutong'yi，ru'guobu'cun'zaizhe'liang'gzi'duan，ze不取zhi值
出院日期：该字段与出院时间同义，如果不存在这两个字段，则不取值
住院天数：该字段与住院日数、住院XX天、共住院字段票据号码：同义，如果不存在这两个字段，则不取值
票据代码：该字段与发票代码同义，如果没有完全对于字段，则取票据代码为左上角特殊颜色（红色、蓝色、黑色、绿色依次优先）大号字体
票据号码：该字段与发票号码同义，如果没有完全对应字段值，则取右上角NO后特殊颜色（红色、蓝色、黑色依次优先）大号字体的数字字符
校验码：取完全对应字段值
开票日期：取右下角的日期
收款单位：取该字段的值
收款人：取该字段的值
复核人：该字段与复核同义，后跟人名
医疗机构类型：取该字段对应的值

"""



prompt_template = """这是一张《{}》的照片，请按序返回下面字段的值：{},要求：请返回下面字段的值,注意：如果图片中未出现该字段，返回‘无’；若图片中出现该字段但没有对应值，返回‘/’。
返回值为：
"""

print(prompt_template.format("《住院小结》", "、".join(["a", "b", "c"])))

image_dir = "/root/autodl-tmp/vlm/dataset/医疗清单发票OCR要素提取任务/大赛1000训练用数据集"


all_image_name = {}
for im_name in os.listdir(image_dir):
    im_name_prefix = im_name.split(".")[0]
    all_image_name[im_name_prefix] = im_name
    


def gen_correct_image_path(image_name:str):
    if image_name.endswith("jpeg") or image_name.endswith("jpg") or image_name.endswith("png"):
        return image_name
    else:
        if image_name in all_image_name:
            return all_image_name[image_name]
        else:
            return None

all_prompt_examples = []

for image_type, image_type_values in all_data_dict.items():

    query = prompt_template.format(image_type, "、".join(imagetype2columns[image_type]))
    for single_image_name, single_image_info_dict in image_type_values.items():
        single_true_image_name = gen_correct_image_path(single_image_name)
        if not single_true_image_name:
            print("---", single_image_name)
            continue
        single_image_path = os.path.join(image_dir, single_true_image_name)
        if not os.path.exists(single_image_path):
            print("===", single_image_path)
            continue
        response = "\n".join([key + ":" + value for key, value in single_image_info_dict.items()])
        all_prompt_examples.append([{"query": query, "response": response, "images":[single_image_path]}, image_type])




print(len(all_prompt_examples))
```

```python
def save_json(data_list, json_path):
    with open(json_path, "w", encoding="utf-8") as fout:
        for example in data_list:
            example_prompt = example[0]
            fout.write(json.dumps(example_prompt, ensure_ascii=False) + "\n")



save_json(train_dataset, "/root/autodl-tmp/vlm/dataset/processed_dataset/yiliaofapiao_ocr_train.jsonl")


save_json(test_dataset, "/root/autodl-tmp/vlm/dataset/processed_dataset/yiliaofapiao_ocr_test.jsonl")

```
